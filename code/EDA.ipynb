```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

```python
conn = sql.connect('../database/bayesatbat.db')
c = conn.cursor()
```

```python
sql_query = pd.read_sql_query("""SELECT * from Altuve""",conn)
```

```python
df =pd.DataFrame(sql_query, columns=['pitch_type','p_throws','launch_speed','launch_angle','release_speed'])
```

```python
def LAtoLS(row,bottomDEG,TopDEG,bottomSpeed,TopSpeed):
    
    if row['launch_speed']>=bottomSpeed and row['launch_speed']<TopSpeed and row['launch_angle']>=bottomDEG and row['launch_angle']<TopDEG:
        val=1
    else:
        val=0
        
    return val
    
    
    
    
    
```

```python
def LA(row,bottomSpeed,TopSpeed):
    if row['launch_speed']>=bottomSpeed and row['launch_speed']<TopSpeed:
        val=1
    else: val=0
    return val
    
    
    
    
    
```

```python
for i in range(0,130,10):
    bottom =i
    top =i+10
    df[f'{bottom} to {top}MPH']=df.apply(LA,args=(bottom,top),axis=1)
    print(bottom,top)
    
```

```python
def launch_angle(row,bottom_angle,Top_angle):
    
    if row['launch_angle']>=bottom_angle and row['launch_angle']<Top_angle:
        val=1
    else: val=0
    return val
    
    
    
```

```python
for i in range(-90,90,10):
    bottom =i
    top =i+10
    df[f'{bottom}to{top}deg']=df.apply(launch_angle,args=(bottom,top),axis=1)
    print(bottom,top)
```
